```python
CHROMA_PATH = "chroma"
DATA_PATH = "data\clinrec"
CHUNK_SIZE = 3000
CHUNK_OVERLAP = 300
```

```python
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import Chroma

from langchain_community.document_loaders import UnstructuredPDFLoader

from langchain_community.document_loaders import PyPDFLoader

import glob 
files = glob.glob(DATA_PATH + r'\*.pdf', recursive=True)
```

```python
loaders = [PyPDFLoader(fp) for fp in files]
empty_documents = []
all_documents = []
count = 0
for loader in loaders:
   count += 1
   print(f'{count}/{len(loaders)}', "Loading raw document..." + loader.file_path)
   raw_documents = loader.load_and_split()
   if len(raw_documents) <1: 
      print('Пустой документ')
      empty_documents.append(loader.file_path)

print('Пустые документны:', empty_documents)
```
